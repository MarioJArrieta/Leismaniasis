## Aplicación de PCA para reducción de dimensionalidad

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# 1. Estandarizar los datos
scaler = StandardScaler()

# Aplicar la estandarización en los datos de entrenamiento
X_train_scaled = scaler.fit_transform(X_train)

# Aplicar la misma estandarización en los datos de prueba
# X_test_scaled = scaler.transform(X_test)

# 2. Inicializar PCA con el número de componentes deseados
k = 10  # Número de componentes principales que deseas mantener
pca = PCA(n_components=k)
pca.fit(X_train_scaled)

# 3. Obtener las varianzas explicadas
explained_variance = pca.explained_variance_ratio_

# Calcular la contribución relativa de cada variable en cada componente principal
contribucion_relativa = explained_variance * 100  # Multiplicado por 100 para expresarlo en porcentaje

# Crear un DataFrame para mostrar las contribuciones relativas
contribucion_df = pd.DataFrame(contribucion_relativa, columns=['Contribución Relativa (%)'], index=[f'PC{i+1}' for i in range(k)])

print("Contribución relativa de las variables en cada componente principal:")
print(contribucion_df)
```

```python
# Obtener las cargas (loadings) de las variables en cada componente principal
loadings = pca.components_

# Calcular las variables con las mayores contribuciones en cada componente
componente_principal_mas_importante = np.argmax(np.abs(loadings), axis=1)
variables_mas_importantes = X_train.columns[componente_principal_mas_importante]
print("Variables con mayor contribución en cada componente principal:")
print(variables_mas_importantes)
```
